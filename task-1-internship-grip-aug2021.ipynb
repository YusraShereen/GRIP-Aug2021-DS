In this task student marks are predicted using linear regression on the basis of study hours. 

### **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

### **Reading Data**

```python
url = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
data = pd.read_csv(url)
data.head()
```

### **Exploring Data**

```python
data.describe()
```

### **Data Cleaning (Checking for null values)**

```python
data.isna().sum()
```

### **Preparing Data**

```python
x = data['Hours'].values.reshape(-1,1)
y = data['Scores'].values.reshape(-1,1)
```

### **Visualizing Data**

```python
data.plot(x = 'Hours',y = 'Scores',kind = 'scatter')
plt.title('Study Hours Vs Students Result')
plt.show()
```

### **Train Test Split**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state=0)
```

### **Training Linear Regression Model**

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_test
```

### **Comparing Actual Vs Predicted**

```python
y_pred = LR.predict(x_test)
df = pd.DataFrame({'Actual': list(y_test), 
                   'Predicted': list(y_pred)})  
df
```

### **Visualizing Regressor on training model**

```python
plt.scatter(x_train,y_train, color = 'red')
plt.plot(x_train, LR.predict(x_train),color = 'blue')
plt.title('Study hours Vs Percentage')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

### **Visualizing Linear regressor fit on test data**

```python
plt.scatter(x_test,y_test, color = 'green')
plt.plot(x_test, y_pred,color = 'blue')
plt.title('Study hours Vs Percentage(on Test Data)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

### **Evaluating Regressor Model**

```python
print('Mean Abs error: ',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared error: ',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Square error: ',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

### **Predicting on different Data**

What will be predicted score if a student studies for 9.25 hrs/ day?

```python
hours = 9.25
h_pred = LR.predict(np.array(hours).reshape(-1,1))
print(h_pred)
```
