### **Problem Statement:**
### Implement an object detector which identifies the classes of the objects in an image

### Using **YOLOv3** pretrained model

### **Loading Libraries**

```python
from imageai.Detection import ObjectDetection
```

### **Creating Object Detector Object**

```python
detector = ObjectDetection()
```

### Giving Path of Input, Output & model to variables

```python
model_path = "/content/drive/MyDrive/Data Science Internship/ObjectDetection/yolo-tiny.h5"
input_path = "/content/drive/MyDrive/Data Science Internship/ObjectDetection/img.JPG"
output_path = "/content/drive/MyDrive/Data Science Internship/ObjectDetection/detected_objects.jpg"
```

### **Using YOLOv3 Pretrained model**

```python
detector.setModelTypeAsTinyYOLOv3()
```

```python
detector.setModelPath(model_path)
```

### **Loading Model**

```python
detector.loadModel()
```

### **Input Image**

```python
from PIL import Image

im = Image.open("/content/drive/MyDrive/Data Science Internship/ObjectDetection/img.JPG")
display(im)
```

### **Detections on Input Image**

```python
detections = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path, minimum_percentage_probability=30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")

```

### **Output Image with Detected Classes**

```python
from PIL import Image

im = Image.open("/content/drive/MyDrive/Data Science Internship/ObjectDetection/detected_objects.jpg")
display(im)
```
